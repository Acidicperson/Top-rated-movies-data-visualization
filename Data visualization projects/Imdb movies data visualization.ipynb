```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

file_path = '/Users/ahmadmayare/Downloads/imdb_top_5000_tv_shows.csv'
tv_shows_df = pd.read_csv(file_path)
top10 = tv_shows_df.sort_values(by='averageRating', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='averageRating', y='primaryTitle', data=top10, palette='mako')
plt.title("Top 10 Highest Rated TV Shows on imdb")
plt.xlabel("averageRating")
plt.ylabel("TV Show")
plt.show
```
