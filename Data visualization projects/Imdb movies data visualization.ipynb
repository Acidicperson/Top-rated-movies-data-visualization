```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


tv_shows_df = pd.read_csv('imdb_top_5000_tv_shows.csv')
top10 = tv_shows_df.sort_values(by='averageRating', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='averageRating', y='primaryTitle', data=top10, palette='mako')
plt.title("Top 10 Highest Rated TV Shows on imdb")
plt.xlabel("averageRating")
plt.ylabel("TV Show")
plt.show
```

```python
from collections import Counter

# Split and count genres
genre_list = tv_shows_df['genres'].dropna().str.split(', ')
flat_genres = [genre for sublist in genre_list for genre in sublist]
genre_counts = Counter(flat_genres).most_common(10)

# Convert to DataFrame
genre_df = pd.DataFrame(genre_counts, columns=['Genre', 'Count'])

# Plot
plt.figure(figsize=(8,5))
sns.barplot(x='Count', y='Genre', data=genre_df, palette='Set2')
plt.title("Top 10 Most Common TV Show Genres")
plt.show()
```
